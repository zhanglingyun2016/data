```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x,y:x*np.exp((-x**2)-(y**2))
x = y = np.linspace(-2.0,2.0,500)  #2-2500
x1,y1 = np.meshgrid(x,y)  #xyxy500
z_f = f(x1,y1)   #z_fxy  
#subplots4--2215055
fig,axes1 = plt.subplots(2,2,dpi=150,figsize=(5,5))

axes1[0,0].imshow(z_f)
axes1[0,1].imshow(z_f,cmap = plt.cm.gray)  #
axes1[1,0].imshow(z_f,cmap = plt.cm.cool)
axes1[1,1].imshow(z_f,cmap = plt.cm.hot)
```
