```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x,y:x*np.exp((-x**2)-(y**2))
x = y = np.linspace(-2.0,2.0,500)  #设置范围，形成一维数组，在2和-2之间分割成500个点
x1,y1 = np.meshgrid(x,y)  #对xy进行网格化处理形成二维数组，对xy重复500次
z_f = f(x1,y1)   #z_f为xy所有函数的结果值  ，用来表示高度的值，所带来不同的颜色
#使用subplots函数，直接形成画布和画框，一个画布分成了4份--》2乘2个画框，150的像素，5乘5的画框大小
fig,axes1 = plt.subplots(2,2,dpi=150,figsize=(5,5))

axes1[0,0].imshow(z_f)
axes1[0,1].imshow(z_f,cmap = plt.cm.gray)  #设置色系
axes1[1,0].imshow(z_f,cmap = plt.cm.cool)
axes1[1,1].imshow(z_f,cmap = plt.cm.hot)
```
