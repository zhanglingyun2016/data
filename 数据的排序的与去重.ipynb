# 数据的排序与去重

```python
import numpy as np
```

```python
# 数据的排序与去重
arr1 = np.random.randint(12,69,(4,6))
arr1
```

```python
# 通过0和1的参数，进行行和列的排序，只是不返回数据，只能进行打印
arr1.sort(0)
arr1
```

```python
arr2 = np.random.randint(12,69,(4,6))
arr2
```

```python
# np.sort()一次性排序，默认的为行排序
res1 = np.sort(arr2,1)
res1
```

```python
# np.sort()一次性排序，默认的为行排序
# 按列排序
res1 = np.sort(arr2,0)
res1
```

### 去重

```python
arr4 = np.random.randint(-36,-11,(7,7))
arr4
```

```python
# 通过广播功能，将0，3，6行进行替换
arr4[[0,3,6]] = np.random.randint(11,36,7)
arr4
```

```python
# 去重操作
# axis=0删除重复的行，axis=1删除重复的列，将第一列进行排序
np.unique(arr4,axis=0)
```

```python
# 列
arr5 = np.random.randint(-36,-11,(6,6))
arr5
```

```python
# 将列进行替换
arr5[:,[0,3,5]] = np.random.randint(10,39,(6,1))
arr5
```

```python
# 删除重复的列，还可以把第一行进行排序
np.unique(arr5,axis=1)
```

```python
# 数据集进行拼接 concatenate()
# 按行进行拼接，要求列要一样
arr1 = np.random.randint(12,36,(5,7))
arr2 = np.random.randint(12,36,(2,7))
arr3 = np.random.randint(12,36,(1,7))
np.concatenate((arr1,arr2,arr3),axis=0)
```

```python
# 数据集进行拼接 concatenate()
# 按列进行拼接，要求行要一样
arr4 = np.random.randint(12,36,(4,3))
arr5 = np.random.randint(12,36,(4,1))
arr6 = np.random.randint(12,36,(4,2))
print(arr4)
np.concatenate((arr4,arr5,arr6),axis=1)
```

```python
# column_stack()列拼接
np.column_stack((arr4,arr5))
```

```python
# row_stack()行拼接
np.row_stack((arr1,arr2))
```

```python
# 按要求进行排序
# 通过函数进行整行整列的排序
arr1 = np.random.randint(-6,8,(7,9))
arr1
```

```python
def arr_sort(arr,num,axis):
    list1 = []
    if axis == 0:                                              # 进行去重操作
        for m1 in [np.where(arr[:,num]==k)[0].tolist() for k in sorted(set(arr1[:,num]))]:
            list1 = list1 + m1
        print(list1)
        return arr[list1]
    else:
        for m2 in [np.where(arr[num]==k)[0].tolist() for k in sorted(set(arr1[num]))]:
            list1 = list1 + m2
        print(list1)
        return arr[:,list1]
        
```

```python
arr_sort(arr1,1,axis=0)
```

```python
# 练习余弦波
t_random = np.random.uniform(0,1,15)
t_random
```

```python
y = lambda t:np.cos(2*np.pi*((t/12)+t_random))
```

```python
fig1,axes1 = plt.subplots(1,1,dpi=150,figsize=(8,4))
dafr_cos.plot(ax=axes1)
```

```python
# 练习3
arr2 = np.random.randint(-6,21,(20,7))
arr2
```

```python
def random_select(arr,num):
    list1 = []
    while len(list1) < num:
        random_num = np.random.randint(0,len(arr),1)[0]
        if random_num not in list1:
            list1.append(random_num)
        else:
            continue
    return arr[[list1]],list1
```

```python
random_select(arr2,3)
```
