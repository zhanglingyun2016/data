# apply和applymap函数

```python
pd1 = pd.DataFrame(np.random.randint(34,189,(500,8)))

pd1.head(5) #只显示5行
f = lambda x : x + 2
print(pd1)
pd1.apply(f,axis = 0)   #作用域 轴线为0或者1的
```

```python
# applymap() 函数
pd2 = pd.DataFrame(np.random.randint(34,189,(500,8)))
func1 = lambda x : x + np.random.randint(16,69,2)
print(func1)
print(pd2)
pd2.applymap(func1)
```

```python
m = [
    [2,3,4],
    [2,4,3]
]
n = [
    [3,4,2],
    [6,2,3]
]

arr1 = np.array(m)
arr2 = np.array(n)
print(arr1)
print("-"*30)
print(arr1)
print("-"*30)
result = arr1 * arr2
print(result)
print("-"*30)
list2 = []
for i in result:
    for j in i:
        list2.append(j)
print(list2)
```

```python
# 生成器

arr1 = [1,2,3,4,5,5]
tou1 = (i**2 for i in arr1)  #返回一个生成器
result1 = list(tou1)
print(result1)  #list遍历

result2 = tuple(arr1)
print(result2)  #tuple遍历

for i in arr1:  #for 循环遍历
    print(i,end=" ")
```
