```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取csv文件

# 使用to_csv（）写入csv文件
fr1 = pd.DataFrame(np.random.randint(12,96,(6,6)),columns = list("abcdef"))
fr1.to_csv("D:\日用文件\数据分析文档\write.csv")
```

```python
#使用read_csv读取write的csv文件
#                          读取csv文件                      分隔形式            具体读取列数
#                   -----------------------------------  =============== -----------------------
read1 = pd.read_csv("D:\日用文件\数据分析文档\write.csv",delimiter = ",",usecols = ["a","b","c"])

read1
```

```python
# converters()函数的使用，字典形式，参数为函数
# converters()函数默认为字符型

read1 = pd.read_csv("D:\日用文件\数据分析文档\write.csv",delimiter = ",",usecols = ["a","b","c"],
                   converters = {"a":lambda x : x*2,"c":lambda y : y+"3"})

read1
```

```python
# converters()函数的使用，字典形式，参数为函数
# converters()函数设置为int或者浮点型

read2 = pd.read_csv("D:\日用文件\数据分析文档\write.csv",delimiter = ",",usecols = ["a","b","c"],
                   converters = {"a":lambda x : int(x)**2,"c":lambda y : float(y)+3})

read2
```

```python
# np.linespace()

data1 = np.linspace(-0.6,0.6,10)

data1
```

```python
# 定义lambda函数func1

func1 = {"a0":lambda x:float(x),"a1":lambda x:float(x)**2,"a2":lambda x:float(x)**3,"a3":lambda x:float(x)**4}

```

```python
# 写入csv文件

dac = pd.DataFrame(np.array([data1,data1,data1,data1]).T,columns = ["a%d" %i for i in range(4)])
dac.to_csv("D:\日用文件\数据分析文档\write3.csv")
```

```python
# 读取csv文件

read1 = pd.read_csv("D:\日用文件\数据分析文档\write3.csv",delimiter = ",",usecols = ["a%d" %i for i in range(4)],
                   converters = func1)

read1
```

```python
# 画图
fig1,asd1 = plt.subplots(1,1,figsize = (4,2),dpi = 150)

for i in range(4):
    asd1.plot(read1["a"+str(i)])

```
