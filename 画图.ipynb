```python
# 创建画布
# import matplotlib as plt
import numpy as np
import matplotlib.pyplot as plt

# figure_1 = plt.figure()  #长宽为4和3  ,dpi为分辨率
figure_1 = plt.figure(1,(4,3),dpi = 100,facecolor = "gray",edgecolor = "yellow",frameon = True)
figure_1
tu1 = figure_1.add_subplot(2,2,1)  #典型图,线图
tu2 = figure_1.add_subplot(2,2,2)  #散点图
tu3 = figure_1.add_subplot(2,2,3)  #柱状图
tu4 = figure_1.add_subplot(2,2,4)  #直方图

x = np.linspace(-3,3,10) # 在-3和3之间分成1000分
f = lambda x: x*np.exp(x)
tu1.plot(x,f(x),"b*--")   #线图

tu2.scatter(np.random.randn(10),np.random.randn(10),c="green",marker="D")   # 点图

x1 = np.random.normal(4,2.1,1000)  #产生的书分布在4位置，2.1为散度，1000个
tu3.hist(x1,bins = 10)  # bins为分为10个柱形。

x2 = ["x%u" %j for j in range(10)]
tu4.bar(x2,height = np.random.randint(2,20,10))   
```

```python
# 饼图
import numpy as np
import matplotlib.pyplot as plt
# 画布与画框
asx,sd = plt.subplots(1,1)
sd.pie(np.arange(4,9),explode = [0,0,0,0,0.1],labels = ["A","B","C","D","E"],
      colors = ["m","r","g","c","b"],autopct = "%.2f%%",pctdistance = 1,shadow = True)
```

```python
# 生成一个两行两列画框的画布，
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),color="k",alpha=0.5)
plt.subplots_adjust(wspace=0.4,hspace=0.2)

```
