```python
# 数据框的索引
import pandas as pd
import numpy as np


# 压缩成一个元组，再转化为数组，进行遍历
arr1 = np.random.randint(2,98,(4,5)).tolist()
dict1 = zip(list("dwfgh"),arr1)

for i in dict1:
    k,v = i
    print(k,v)

arr2 = np.random.randint(1,9,9).tolist()
for i in range(3):
    for j in arr2:
        j = 0
print(arr2)
    
```

```python
arr3 = pd.Series([25,36,24],index=list("ghd"))
arr3
```

```python
# 通过字典创建数据框，字典的键做为列索引，字典的index作为数据框的行索引
dict1 = {"A1":pd.Series([1,2,3],index=list("asd")),
        "A2":pd.Series([111,222,333],index=list("asd")),
        "A3":pd.Series([1,22,33],index=list("asd"))}

# print(dict1)
pd.DataFrame(dict1)    # 通过字典创建数据框
```

```python
# 下标值进行重置
ser3= pd.Series([12,66,99,78,6,47],index=list("fedcba"))
print(ser3)
print("-"*50)

# 下边重置后的
result = ser3.reindex(list("abcdef"))
print(result)
print("-"*50)

# 对下标值进行添加和减少操作   添加了一个g的索引，赋值为99999
result1 = ser3.reindex(list("abcdefg"),fill_value=99999)
print(result1)
print("-"*50)

# 对下标值进行减少操作
result2 = ser3.reindex(list("abc"))
print(result2)
```

```python
# 补值
ser7 = pd.Series(np.random.randint(1,150,50))
for i in range(2):   # 随机产生两个下标值
    k = ser7.index[np.random.randint(0,49,1)]
    ser7[k] = np.nan   #取出下表值进行赋两次为nan

ser7
```

```python
# 删除索引值
ser7.drop([13,20])
```

```python
# 删除索引后进行补值
ser7.reindex(range(50),method="nearest")   #根据附近补值
```

```python
# 可以直接索引列，不可以直接索引行
pd1 = pd.DataFrame(np.random.randint(1,56,(5,5)),columns = list("ABCDE"),index = list("abcde"))
print(pd1)
print(pd1["A"] ) #直接索引列

print(pd1.loc["a"] ) #加入。loc 索引行

pd1.loc["a","A"]  #使用loc["行值","列值"]，查询数值
```

```python
# 选区
pd1 = pd.DataFrame(np.random.randint(1,56,(5,5)),columns = list("ABCDE"),index = list("abcde"))
result = pd1.loc["a":"c","B":"E"]   #选区操作
print(result)
pd1
```

```python
# 行式花式索引
pd1 = pd.DataFrame(np.random.randint(1,56,(5,5)),columns = list("ABCDE"),index = list("abcde"))

pd1.loc[["a","c","e"],:]
```

```python
# 列式花式索引
pd2 = pd.DataFrame(np.random.randint(1,56,(5,5)),columns = list("ABCDE"),index = list("abcde"))

pd2.loc[:,["A","C","E"]]
```

```python
# 花式索引整合
pd3 = pd.DataFrame(np.random.randint(1,56,(5,5)),columns = list("ABCDE"),index = list("abcde"))

pd3.loc[["a","c","e"],["A","C","E"]]
```

## 数据框的切片与赋值

```python
# 切片
pd5 = pd.DataFrame(np.random.randint(1,56,(5,5)),columns = list("ABCDE"),index = list("abcde"))
pd6 = pd5.copy()
pd6["a":"c"]
pd6
```

```python
# 赋值操作
pd6 = pd.DataFrame(np.random.randint(1,56,(5,5)),columns = list("ABCDE"),index = list("abcde"))

pd7 = pd6.copy()
pd7.loc["a":"b","A":"B"] = 999  #进行赋值操作
pd7
```

```python
# 对指定位置赋值-某某某、

pd7 = pd.DataFrame(np.random.randint(1,56,(5,5)),columns = list("ABCDE"),index = list("abcde"))

# 通过字典套字典对指定位置赋值
pd7.loc["a":"b","A":"B"] = pd.DataFrame({"A":{"a":111,"b":444},"B":{"a":222,"b":777}})
pd7
```
