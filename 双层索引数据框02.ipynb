```python
# 重复的内部的行索引聚合计算

import numpy as np
import pandas as pd

arr = pd.DataFrame(np.random.randint(3,65,(6,4)),
              index = [["morning","morning","morning","afternoon","afternoon","afternoon"],
                       ["r"+str(i) for i in range(2,5)]+["r"+str(i) for i in range(3,6)]],
              columns = [["A","A","B","B"],["a","b","c","d"]])

arr
```

```python
# 聚合计算  ,将重复的r3和r4合并值
     
arr.sum(0,level=-1)
```
