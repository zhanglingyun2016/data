###### 练习1，及apply()函数的使用

```python
# 最大值最小值归一化
# 数组中的某值减去最小值，除以数组最大值减去最小值，且结果总是大于0小于1
maxmin_func = lambda x:(x-x.min())/(x.max()-x.min())

# Z_score标准化函数
# 数组中的某值减去平均值,除以某值的标准差
z_score = lambda y:(y-y.mean())/(y.std())
```

```python
pan1 = pd.DataFrame(np.random.randint(-121,254,(10,6)),columns=list("abcdef"))
pan1
```

```python
# 最大值最小值归一化
# apply()函数是将定义的某个函数，用到数据框中的某一行某一列
# axis=0按列进行计算，axis=1按行进行计算，默认为0
# 多数按列计算，一列算是一个变量
pan1.apply(maxmin_func,axis=0)
```

```python
# z_score标准化
# apply()函数是将定义的某个函数，用到数据框中的某一行某一列
# axis=0按列进行计算，axis=1按行进行计算，默认为0
# 多数按列计算，一列算是一个变量
pan1.apply(z_score,axis=0)
```
