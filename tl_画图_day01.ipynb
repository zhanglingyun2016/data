# 图灵课程画图

```python
import matplotlib.pyplot as plt # 画图
from mpl_toolkits.mplot3d import Axes3D #将画布进行3D化
import numpy as np
```

```python
# 克莱因瓶
# 公式
def fun(u1,u2):
    return 2.6*(1-np.sin(u1))*np.cos(u1)+(2-np.cos(u1))*np.cos(u2)*(2*np.exp(-((u1/2-np.pi)**2))-1),\
            (2-np.cos(u1))*np.sin(u2),\
            9*np.sin(u1)+0.5*(2-np.cos(u1))*np.sin(u1)*np.cos(u2)*np.exp(-((u1-1.5*np.pi))**2)
```

```python
# 先生成画布，参数 ：      编号,分辨率 ， 画布大小
fig1 = plt.figure(0,dpi=100,figsize=(5,8))
ax = Axes3D(fig1)

# 分割为1千个点
u1 = np.linspace(0,2*np.pi,1000)
u2 = np.linspace(0,2*np.pi,1000)
U1,U2 = np.meshgrid(u1,u2)  # 进行网格化
z0,z1,z2 = fun(U1,U2)
# rstride采样，每隔20采一次，不显得密集，20行20列
ax.plot_wireframe(z0,z1,z2,rstride=20,cstride=20,color="k")

# 仰角 方位角
ax.view_init(elev=20,azim=265)
```

```python
# 势图
# 函数生成公式为F(x,y,f(x,y))
fig2 = plt.figure(0,dpi=150,figsize=(4,4))
ax = Axes3D(fig2) # 3D化
f = lambda x,y:x*np.exp(-(x**2+y**2))
x = np.linspace(-5,5,3000)
y = np.linspace(-5,5,3000)
x1,y1 = np.meshgrid(x,y)
z = f(x1,y1)
ax.plot_surface(x1,y1,z,cmap="Spectral_r")
```

```python
# 色系
def fun(x,y):
    return x*np.exp(-(x**2+y**2))
fig3 = plt.figure(0,dpi=150,figsize=(4,4))
x = y = np.linspace(-2,2,500)
x1,y1 = np.meshgrid(x,y)
z = fun(x1,y1)

# 直接生成画布和画框，但无法生成三维视图
fig3,axes1 = plt.subplots(2,2,dpi=150,figsize=(5,5))
axes1[0,0].imshow(z)
axes1[0,1].imshow(z,cmap=plt.cm.gray)
axes1[1,0].imshow(z,cmap=plt.cm.cool)
axes1[1,1].imshow(z,cmap=plt.cm.hot)
```
